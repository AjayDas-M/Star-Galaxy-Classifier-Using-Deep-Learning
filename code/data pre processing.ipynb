```python
import os
import shutil
from sklearn.model_selection import train_test_split
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Paths to the original dataset
data_dir = '/home/user/ajay das/mini-project/data/'
star_dir = os.path.join(data_dir, 'star')
galaxy_dir = os.path.join(data_dir, 'galaxy')

# Paths to save the split dataset
output_dir = '/content/drive/MyDrive/trial'
train_dir = os.path.join(output_dir, 'train')
test_dir = os.path.join(output_dir, 'test')

# Creating the output directories
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Creating subdirectories for each class
os.makedirs(os.path.join(train_dir, 'star'), exist_ok=True)
os.makedirs(os.path.join(train_dir, 'galaxy'), exist_ok=True)
os.makedirs(os.path.join(test_dir, 'star'), exist_ok=True)
os.makedirs(os.path.join(test_dir, 'galaxy'), exist_ok=True)
```

```python
def split_dataset(class_dir, train_size=0.8):
    images = os.listdir(class_dir)
    train_images, test_images = train_test_split(images, train_size=train_size, random_state=42)
    return train_images, test_images

def move_images(images, source_dir, dest_dir):
    for image in images:
        source_path = os.path.join(source_dir, image)
        dest_path = os.path.join(dest_dir, image)
        shutil.move(source_path, dest_path)

# Split and move star images
train_images, test_images = split_dataset(star_dir)
move_images(train_images, star_dir, os.path.join(train_dir, 'star'))
move_images(test_images, star_dir, os.path.join(test_dir, 'star'))

# Split and move galaxy images
train_images, test_images = split_dataset(galaxy_dir)
move_images(train_images, galaxy_dir, os.path.join(train_dir, 'galaxy'))
move_images(test_images, galaxy_dir, os.path.join(test_dir, 'galaxy'))

print("Dataset split completed.")
```

```python
def split_dataset(class_dir, train_size=0.8):
    images = os.listdir(class_dir)
    train_images, test_images = train_test_split(images, train_size=train_size, random_state=42)
    return train_images, test_images

def copy_images(images, source_dir, dest_dir):
    for image in images:
        source_path = os.path.join(source_dir, image)
        dest_path = os.path.join(dest_dir, image)
        shutil.copy(source_path, dest_path)

# Split and copy star images
train_images, test_images = split_dataset(star_dir)
copy_images(train_images, star_dir, os.path.join(train_dir, 'star'))
copy_images(test_images, star_dir, os.path.join(test_dir, 'star'))

# Split and copy galaxy images
train_images, test_images = split_dataset(galaxy_dir)
copy_images(train_images, galaxy_dir, os.path.join(train_dir, 'galaxy'))
copy_images(test_images, galaxy_dir, os.path.join(test_dir, 'galaxy'))

print("Dataset split and copy completed.")
```
